## 식당명(MCT_NM)의 unique한 값만 가진 데이터 프레임 만들기

```python
import pandas as pd
```

```python
# 데이터 프레임 불러오기
df = pd.read_csv("../data/data_prep_with_avg_temp_final.csv", encoding='utf8')
df
```

```python
df.columns
```

```python
df_unique_add = df.drop_duplicates(subset=['MCT_NM', 'ADDR'])
df_unique_add.shape
```

```python
df_unique_loc = df.drop_duplicates(subset=['MCT_NM', 'latitude', 'longitude'])
df_unique_loc.shape
```

```python
df_unique_rest = df_unique_loc[['MCT_NM', 'area', 'MCT_TYPE', 'ADDR', 'latitude', 'longitude']]
df_unique_rest
```

```python
# area, MCT_TYPE, MCT_NM 기준으로 정렬하고 인덱스 리셋
df_unique_rest_sorted = df_unique_rest.sort_values(by=['area', 'MCT_TYPE', 'MCT_NM']).reset_index(drop=True)

# 결과 출력
df_unique_rest_sorted
```

```python
# 데이터프레임 저장
df_unique_rest_sorted.to_csv("../data/unique_restaurant.csv", encoding='cp949')
```
